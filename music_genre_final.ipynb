```python
spotify_columns = ['acousticness', 'danceability', 'energy','instrumentalness', 'liveness', 
                   'loudness','speechiness', 'valence']
```

Spotify присваивает восемь качеств каждой дорожке. Большинство из них имеют значения от 0,0 (не имеет этого качества) до 1,0 (абсолютно имеет это качество). 

1. Акустичность (`acousticness`): показывает, насколько Spotify уверен в том, что трек был исполнен акустически (без электронных инструментов). 
2. Танцевальность (`danceability`): в документации Spotify «танцевальность» описывается как «насколько подходит трек для танцев на основе комбинации музыкальных элементов, включая темп, стабильность ритма, силу удара и общую регулярность.
3. Энергия (`energy`): используя несколько различных качеств, таких как темп, громкость и общую активность в песне, Spotify составляет «энергетическую» оценку, которая показывает, насколько вероятно, что трек разбудит вас утром. 
4. Инструментальность (`instrumentalness`): Spotify также предсказывает, насколько вероятно, что песня будет без вокала, не считая звуков во рту, таких как «ох» и «ааа». 
5. Живучесть (`liveness`): прислушиваясь к шуму толпы, Spotify сообщает нам, была ли песня записана перед живой студийной аудиторией в стиле Full House.
6. Громкость (`loudness`): хотя вы можете вручную отрегулировать громкость в любое время, Loudness сообщает нам об относительной громкости по всей длине песни (в основном, вам нужно увеличить громкость, чтобы услышать эту песню? ). Громкость измеряется от -60 до 0 децибел, при этом средняя громкость большинства песен составляет около - 5 дБ. 
7. Речь (`speechiness`): противоположность инструментальности, измеряет присутствие произнесенных слов на треке. Подкасты, вероятно, будут иметь высокие оценки.
8. Валенс (`valence`): Конечно, музыка хорошая , но счастливая ли она? Валентность использует такие меры, как тональность и темп, чтобы определить, есть ли в песне положительные или отрицательные эмоции.

#### Кодирование и стандартизация признаков

***Масштабирование признаков*** - процесс приведения всех признаков или независимых переменных (переменных, отличных от целевой переменной) практически к одному масштабу, чтобы каждый признак был одинаково важен.

Один из методов масштабирования — стандартизация данных. В машинном обучении это метод, в котором значения центрируются вокруг среднего с единичным стандартным отклонением (µ = 0 и σ = 1). Это означает, что атрибуты среднего значения равны нулю, а результирующее распределение имеет стандартное отклонение единицы.

В рамках данного проекта будем использовать StandardScaler().

Для работы с признаками надо произвести ***кодирование категориальных признаков*** - процедуру, которая представляет собой некоторое преобразование категориальных признаков в численное представление. 

Кодирование данных будем выполнять с помощью инструмента OrdinalEncoder:
1. для столбца key - расположив музыкальные ключи в порядке их высотности звучания,
2. для столбца date - расположив даты в порядке их возрастания,
3. для столбца mode - от минора к мажору.

```python
ordinal_key = OrdinalEncoder(categories=[['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A','A#', 'B']], 
                             handle_unknown ='use_encoded_value',
                             unknown_value= np.nan)

ordinal_date = OrdinalEncoder(categories=[['1-Apr','3-Apr', '4-Apr', '5-Apr']])

ordinal_mode = OrdinalEncoder(categories=[['Major', 'Minor']], 
                             handle_unknown ='use_encoded_value',
                             unknown_value= np.nan)
```

#### Заполнение пропусков

Заменим аномальные значения в столбце duration_s на пропуски, для того, чтобы была возможность их впослествии заполнить с помощью встроенным трансформатором из библиотеки scikit-learn.

```python
features_train.loc[features_train['duration_s'] < 0, 'duration_s'] = np.nan
features_test.loc[features_test['duration_s'] < 0, 'duration_s'] = np.nan
df_test.loc[df_test['duration_s'] < 0, 'duration_s'] = np.nan
```

Для заполнения количественных признаков используем SimpleImputer, для категориальных признаков - IterativeImputer.

#### Создание препроцессора

##### Для модели Случайного леса

```python
num_pipeline = make_imblearn_pipeline(SimpleImputer(strategy='median'))

key_pipeline = make_imblearn_pipeline(SimpleImputer(strategy = 'most_frequent'), 
                                      ordinal_key)

mode_pipeline = make_imblearn_pipeline(SimpleImputer(strategy = 'most_frequent'),
                                      ordinal_mode)

date_pipeline = make_imblearn_pipeline(ordinal_date)
```

Создадим preprocessor, объекта класса ColumnTransformer из библиотеки scikit-learn для обработки и преобразования разных типов признаков в наборе данных перед их подачей на модель машинного обучения.

```python
preprocessor_rf = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_columns),  
        ('key', key_pipeline, key_colunmns), 
        ('date', date_pipeline, date_columns),
        ('mode', mode_pipeline, mode_colunmns)
    ], remainder='passthrough')
```

##### Для модели градиентного бустинга

```python
num_pipeline = make_imblearn_pipeline(SimpleImputer(strategy='median'))

key_pipeline = make_imblearn_pipeline(SimpleImputer(strategy = 'most_frequent'))

mode_pipeline = make_imblearn_pipeline(SimpleImputer(strategy = 'most_frequent'))
```

```python
preprocessor_boost = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_columns),  
        ('key', key_pipeline, key_colunmns), 
        ('mode', mode_pipeline, mode_colunmns)
    ], remainder='passthrough')
```

### Определение метрик качества модели

В задачах машинного обучения для оценки качества моделей и сравнения различных алгоритмов используются метрики, а их выбор осуществляется, прежде всего исходя из самих данных, а также поставленной цели.
Существуют следующие метрики качества моделей для решения задачи мультиклассовой классификации:

***1. Матрица ошибок (Сonfussion matrix)***

Матрица ошибок - матрица, которая показывает количество верно и неверно классифицированных примеров для каждого класса. Из матрицы неточностей можно вычислить различные метрики, такие как точность, полнота и F1-метрика.


***2. Accuracy (Доля правильных ответов)***

Доля правильно классифицированных примеров (всех классов) относительно общего числа примеров. Она показывает, насколько хорошо модель классифицирует данные в целом. Однако точность может быть вводящей в заблуждение, особенно при несбалансированных классах.


***3. Preсision (Точность)***

Точность определяет, как много отрицательных ответов нашла модель, пока искала положительные. Precision представляет собой отношение числа семплов, верно классифицированных как Positive, к общему числу выборок с меткой Positive (распознанных правильно и неправильно). 


***4. Recall (Полнота)***

Полнота выделяет, какую долю положительных ответов среди всех ответов выделила модель. Recall рассчитывается как отношение числа Positive выборок, корректно классифицированных как Positive, к общему количеству Positive семплов. Recall измеряет способность модели обнаруживать выборки, относящиеся к классу Positive. Чем выше recall, тем больше Positive семплов было найдено.


***5. F1 (F1-мера)***

F1-мера - это среднегармоническое полноты и точности. 

***Исходя из задачи поставленной в рамках данного проекта, основной метрикой явлется F1-micro.*** 

F1-micro - это один из подходов, при котором усредняется точность (precision) и полнота (recall) для всех классов, а затем вычисляется F1-метрика.
F1-micro усредняет точность и полноту по всем классам путем учета общего числа истинно положительных, ложно положительных и ложно отрицательных примеров для всех классов.
